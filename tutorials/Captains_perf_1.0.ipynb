```python
odis["Won_pc"] = odis.apply( lambda rec: ( rec.Won / rec.Matches )
                            * 100, axis = 1 )
```

```python
odis["Won_pc1"] = odis["Won"] / odis["Matches"]
```

```python
odis.head()
```

```python
odis["Won_pc"] = odis.apply( lambda rec: np.round( ( rec.Won / rec.Matches ), 4 ) 
                            * 100, axis = 1 )
```

```python
odis.head()
```

```python
odis["Lost_pc"] = odis.apply( lambda rec: np.round( ( rec.Lost / rec.Matches ), 4 ) 
                             * 100, axis = 1 )

odis["Toss_pc"] = odis.apply( lambda rec: np.round( ( rec.Toss / rec.Matches ), 4 ) 
                             * 100, axis = 1 )
```

```python
odis.head()
```

```python
odis.sort( ["Won_pc"], ascending = False )[:10]
```

```python
## Let's consider only the captains that have played atleast 100 matches....
odis100 = odis[odis.Matches > 50 ]
```
