```python
import requests
from bs4 import BeautifulSoup
```

```python
url ="http://www.amazon.com/Oneplus-Unlocked-Smartphone-Android-Nano-SIM/product-reviews/B013WLT63A/ref=cm_cr_dp_see_all_summary?ie=UTF8&showViewpoints=1&sortBy=byRankDescending"
```

```python
review_html = requests.get(url).text
```

```python
review_html
```

```python
soup = BeautifulSoup(review_html)
review_texts = soup.findAll('span', {'class': 'a-size-base review-text'})
```

```python
len( review_texts )
```

```python
review_titles = soup.findAll('a', {'class': 'a-size-base a-link-normal review-title a-color-base a-text-bold'})
```

```python
len( review_titles )
```

```python
for review_title in review_titles:
    print( review_title.text )
```
